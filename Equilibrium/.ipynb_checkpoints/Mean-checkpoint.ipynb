```python
x =  np.linspace(-5,5,100)

def func(x,B):
    return  3*(x**2-1)*(x-1)*(x/B+1)
```

```python
plt.plot(x, func(x,1),'k',label = '$B=1$')
plt.plot(x, func(x,1.5),'c',label = '$B=1.5$')
#plt.plot(x, func(x,0.5),'b',label = '$B=0.5$')
plt.plot(x, func(x,2),'r',label = '$B=2$')
#plt.plot(x, func(x,5),'m',label = '$B=5$')
#plt.plot(x, func(x,0.1),'c',label = '$B=0.1$')
#plt.plot(x, func(x,3),'g',label = '$B=3$')
plt.legend()

plt.ylim(-4,4)
```

```python
plt.plot(x, func(x,1),'k',label = '$B=1$')
plt.plot(x, func(x,0.9),'b',label = '$B=0.9$')
plt.plot(x, func(x,0.7),'r',label = '$B=0.7$')
plt.plot(x, func(x,0.5),'y',label = '$B=0.5$')
plt.plot(x, func(x,0.3),'m',label = '$B=0.3$')
plt.plot(x, func(x,0.1),'c',label = '$B=0.1$')

plt.legend()

plt.ylim(-16,20)
```

```python
plt.plot(x, func(x,1),'k',label = '$B=1$')
plt.plot(x, func(x,1.5),'y',label = '$B=1.5$')
plt.plot(x, func(x,2),'b',label = '$B=2$')
plt.plot(x, func(x,3),'r',label = '$B=3$')

plt.legend()

plt.ylim(-20,20)
```

```python
MinR = 1
def MinL(B):
    return  -((3*B)/8)-(np.sqrt(9*B**2-10*B+17)/8)-1/8

def Max(B):
    return  -((3*B)/8)+(np.sqrt(9*B**2-10*B+17)/8)-1/8
```

```python
B=2
U2R = func(Max(B),B)-func(MinR,B)
U2L = func(Max(B),B)-func(MinL(B),B)
print(U2R)
print(U2L)
```

```python
B=1
U1R = func(Max(B),B)-func(MinR,B)
U1L = func(Max(B),B)-func(MinL(B),B)
print(U1R)
print(U1L)
```
